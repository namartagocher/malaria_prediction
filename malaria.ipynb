```python
values = {"fever":{"no":0, "yes":1},"cold":{"no":0, "yes":1},"rigor":{"no":0, "yes":1},"fatigue":{"no":0, "yes":1},
          "headace":{"no":0, "yes":1},"bitter_tongue":{"no":0, "yes":1},"vomitting":{"no":0, "yes":1},
          "diarrhea":{"no":0, "yes":1},"Convulsion":{"no":0, "yes":1},"Anemia":{"no":0, "yes":1},"jundice":{"no":0, "yes":1},
          "cocacola_urine":{"no":0, "yes":1},"hypoglycemia":{"no":0, "yes":1},"prostraction":{"no":0, "yes":1},
          "hyperpyrexia":{"no":0, "yes":1},"hyperpyrexia":{"no":0, "yes":1},"severe_maleria":{"no":0, "yes":1}}

malaria.replace(values,inplace=True)
```

```python
malaria.head()
```

```python
y = malaria['severe_maleria']
```

```python
y
```

```python
new_malaria = malaria.drop('severe_maleria', axis=1)
```

```python
new_malaria.head()
```

```python
new_malaria = pd.get_dummies(new_malaria)
```

```python
new_malaria = new_malaria.drop('age', axis=1)
```

```python
new_malaria.fillna(0, inplace=True)
```

```python
new_malaria.head()
```

```python
print(malaria.groupby('severe_maleria').size())
```

```python
import seaborn as sns

sns.countplot(malaria['severe_maleria'],label="Count")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(new_malaria, y, test_size=0.25, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier

training_accuracy = []
test_accuracy = []
# try n_neighbors from 1 to 10
neighbors_settings = range(1, 11)

for n_neighbors in neighbors_settings:
    # build the model
    knn = KNeighborsClassifier(n_neighbors=n_neighbors)
    knn.fit(X_train, y_train)
    # record training set accuracy
    training_accuracy.append(knn.score(X_train, y_train))
    # record test set accuracy
    test_accuracy.append(knn.score(X_test, y_test))

plt.plot(neighbors_settings, training_accuracy, label="training accuracy")
plt.plot(neighbors_settings, test_accuracy, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
plt.savefig('knn_compare_model')
```

```python
y1=[]
y2=[]
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train, y_train)
y1.append(knn.score(X_train, y_train))
y2.append(knn.score(X_test, y_test))
print('Accuracy of K-NN classifier on training set: {:.2f}'.format(knn.score(X_train, y_train)))
print('Accuracy of K-NN classifier on test set: {:.2f}'.format(knn.score(X_test, y_test)))
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(C=1).fit(X_train, y_train)
y1.append(logreg.score(X_train, y_train))
y2.append(logreg.score(X_test, y_test))
print("Training set accuracy: {:.3f}".format(logreg.score(X_train, y_train)))
print("Test set accuracy: {:.3f}".format(logreg.score(X_test, y_test)))
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)
y1.append(tree.score(X_train, y_train))
y2.append(tree.score(X_test, y_test))
print("Accuracy on training set: {:.3f}".format(tree.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(X_train, y_train)
y1.append(svc.score(X_train, y_train))
y2.append(svc.score(X_test, y_test))
print("Accuracy on training set: {:.2f}".format(svc.score(X_train, y_train)))
print("Accuracy on test set: {:.2f}".format(svc.score(X_test, y_test)))
```

```python
x=['knn','LogisticRegression','DecisionTree','svm']
plt.figure()
pos=np.arange(len(x))
bars=plt.bar(pos,y1,align='center',linewidth=0,color='lightslategrey')
bars[0].set_color('#1F77B4')

plt.xticks(pos,x)

plt.title('training accuracy',alpha=0.8)

for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.show()
```

```python
plt.figure()
pos=np.arange(len(x))
bars=plt.bar(pos,y2,align='center',linewidth=0,color='lightslategrey')
bars[0].set_color('#1F77B4')

plt.xticks(pos,x)

plt.title('testing accuracy',alpha=0.8)

for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.show()
```
